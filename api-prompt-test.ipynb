```python
from openai import OpenAI

# 1. 配置你的客户端
# 注意：虽然是 OpenAI 库，但我们连的是国内厂商的服务器
client = OpenAI(
    api_key="sk-xxxxxxxxxxxxxxx",
    base_url="https://llmapi.paratera.com"  # 千万别填成了 openai.com，要填 Chatbox 里那个
)

# 2. 构造对话历史
# 这里的 messages 列表，就是 Chatbox 后台实际发送给模型的东西
messages = [
    {"role": "system", "content": "你是一个资深的 Python 助教，回答要简洁。"},
    {"role": "user", "content": "请用一句话解释什么是递归？"}
]

print("正在发送请求...")

# 3. 发送请求，completions代表续写的意思，从返回的大json对象中取出文本回复传给response
response = client.chat.completions.create(
    model="Qwen3-235B-A22B-Instruct-2507",  # 如 qwen3-xxx 或 glm-4
    messages=messages,
    stream=False  # 先关掉流式，方便看结果
    # temperature=0.7 # 你可以在这里把玩 Chatbox 里的温度参数
)

# 4. 打印结果
print("模型回复：")
print(response.choices[0].message.content)
```

```python
from openai import OpenAI
import json

client = OpenAI(api_key="sk-xxxxxxxxxxxxxxx", base_url="https://llmapi.paratera.com")

def translate_text(text, target_lang="English"):
    """
    这是一个封装好的功能函数，
    输入是文本，输出是翻译结果。
    """
    prompt = f"请将以下文本翻译成 {target_lang}，只输出翻译后的内容，不要有解释：\n{text}"
    # f允许字符串内通过{}填入变量的值
    response = client.chat.completions.create(
        model="Qwen3-235B-A22B-Instruct-2507",
        messages=[
            {"role": "system", "content": "你是一个专业的翻译引擎。"},
            {"role": "user", "content": prompt}

        ],
        n = 2
    )
    return response.choices[1].message.content  #尝试让大模型给两个输出并返回第二个输出

# 模拟用户调用
if __name__ == "__main__":
    user_input = input("请输入你想翻译的中文：")
    result = translate_text(user_input, "English")
    print(f"--------\n翻译结果：{result}")
```

```python
from openai import OpenAI

# 1. 配置 (填你自己的 Key 和 URL)
client = OpenAI(api_key="sk-xxxxxxxxxxxxxxx", base_url="https://llmapi.paratera.com")


# 2. 初始化一个空的对话历史列表
# 系统人设只用加一次
messages = [
    {"role": "system", "content": "你是一个有用的助手。"}
]

print("--- 开始对话 (输入 'quit' 退出) ---")

while True:
    # 3. 获取你的输入
    user_input = input("User: ")

    if user_input.lower() == "quit":
        break

    # 4. 【关键步骤】把你的问题加入历史
    messages.append({"role": "user", "content": user_input})

    # 5. 发送请求 (把整个 messages 列表发过去)
    response = client.chat.completions.create(
        model="Qwen3-235B-A22B-Instruct-2507", # 比如 qwen3-xxx
        messages=messages,
        stream=False
    )

    # 6. 获取 AI 的回复
    ai_reply = response.choices[0].message.content
    print(f"Assistant: {ai_reply}\n")

    # 7. 【关键步骤】把 AI 的回复也加入历史！
    # 这样下一次循环时，模型就能看到自己刚才说过的话了
    messages.append({"role": "assistant", "content": ai_reply})

    # (可选) 打印一下现在的历史列表长度，让你感受一下 token 是怎么变多的
    # print(f"[当前上下文包含 {len(messages)} 条消息]")
```
